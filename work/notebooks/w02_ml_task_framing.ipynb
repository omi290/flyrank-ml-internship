# 4. Unit of Analysis

Each row in the dataset represents **one webpage**.

For every webpage, the dataset contains historical search metrics such as impressions, clicks, CTR, and average search position.

The machine learning model will make one prediction per webpage.

```python
df["priority"] = (
    (df["ctr"] < df["ctr"].median()) &
    (df["avg_position"] > df["avg_position"].median())
)

df["priority"] = df["priority"].map({True: "High Priority", False: "Low Priority"})

df[["ctr", "avg_position", "priority"]].head()
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*



A fixed rule such as "refresh every page with CTR below 2%" ignores many other factors that influence search performance.

Machine learning can learn relationships among multiple metrics, including impressions, CTR, clicks, and average position.

By considering these variables together, an ML model can produce more informed recommendations than a single manually chosen threshold.

The model is intended to support decision-making rather than replace human judgment.
