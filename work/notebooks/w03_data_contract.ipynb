```python
from huggingface_hub import hf_hub_download

march_path = hf_hub_download(
    repo_id="FlyRank/internship-warehouse",
    repo_type="dataset",
    filename="fact_content_daily_performance/month=2026-03/data_0.parquet",
    token=HF_TOKEN,
)

print(march_path)
```

```python
import duckdb

db = duckdb.connect()

db.sql(f"""
CREATE OR REPLACE VIEW march_data AS
SELECT *
FROM read_parquet('{march_path}')
""")
```

```python
db.sql("""
DESCRIBE march_data
""").df()
```




Query 1 — Verify Grain


```python
db.sql("""
SELECT
    report_date,
    client_hash_id,
    content_hash_id,
    COUNT(*) AS cnt
FROM march_data
GROUP BY
    report_date,
    client_hash_id,
    content_hash_id
HAVING COUNT(*) > 1
LIMIT 5;
""").df()
```

Query 2 — Row Count + Date Span

```python
db.sql("""
SELECT
    COUNT(*) AS total_rows,
    MIN(report_date) AS start_date,
    MAX(report_date) AS end_date
FROM march_data;
""").df()
```

Query 3 — Availability Check

```python
db.sql("""
SELECT
    COUNT(*) AS rows_with_ga4
FROM march_data
WHERE ga4_data_available IS TRUE;
""").df()
```

## Verification

The unit of analysis is one content page for one client on one report date.

The verification queries confirm:

- No duplicate rows exist for the chosen grain.
- The selected data covers the March 2026 reporting window.
- Only rows where `ga4_data_available IS TRUE` are used for GA4-based analysis.

Five Features

```python
db.sql("""
SELECT
    gsc_impressions,
    gsc_clicks,
    gsc_avg_position,
    ga4_pageviews,
    ga4_sessions
FROM march_data
WHERE ga4_data_available IS TRUE
LIMIT 20;
""").df()
```

## Feature Frame

| Feature | Available when? |
|----------|-----------------|
| gsc_impressions | Known before making a prediction because it comes from historical Search Console data. |
| gsc_clicks | Available from historical search performance before prediction time. |
| gsc_avg_position | Historical ranking signal known before prediction. |
| ga4_pageviews | Historical website traffic already observed before prediction. |
| ga4_sessions | Historical GA4 engagement available before prediction. |
